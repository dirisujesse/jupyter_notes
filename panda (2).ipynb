```python
import csv
```

```python
f = open("us_marriages_divorces.csv")
years = []
divorces = []
for i in csv.reader(f):
    years.append(i[0])
    divorces.append(i[2])
    
del years[0]
del divorces[0]

x = years
y = divorces

w = []
z = []
for i in x:
    inter = int(i) if i.isdigit() else str(i)
    z.append(inter)

for j in y:
    nu = int(j) if j.isdigit() else None
    w.append(nu)
    
w
```

```python
import pylab as pl
```

```python
pl.title("Divorce Statistics in America Since the 19th Century")
pl.xlabel("Years")
pl.ylabel("Divorce Rates")
pl.plot(z , w,'c')
pl.savefig("output.pdf", dpi=600, format="pdf")
pl.show()
```
