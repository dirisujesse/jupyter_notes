#### Labeling Graphs

```python
x = [6,7,8,9,10]
y = [2,4,1,5,3]
y2 = [4,2,5,3,1]
plt.plot(x, y, 'c', label='First Plot')
plt.plot(x, y2, 'm--', label='Second Plot')
#plt.yticks(range(len(y)))
plt.xlabel('Plot Label')
plt.ylabel('Important Var')
plt.title('Nice Graph')
plt.legend(loc='best')
plt.show()
```

#### Barcharts and Histograms

```python
x = [2,4,6,8,10]
y = [6,4,9,3,7]

plt.bar(x, y, color='#ccd231', label='Women')
plt.bar(x, y2, color='#265fec', label='Children')
plt.ylim(0, 9.2)
plt.xlim(1.8, 11)
plt.xlabel('Plot Label')
plt.ylabel('Important Var')
plt.title('Nice Graph')
plt.legend(loc='best')
plt.show()
```

```python
population_ages = []

for i in range(130):
    rand = random.randint(1, 130)
    population_ages.append(rand)
print(population_ages)
```

```python
bins = [x for x in range(0, 131, 10)]

plt.hist(population_ages, bins, histtype='bar', color='c', label='First Plot', rwidth=0.9)
plt.xlabel('Plot Label')
plt.ylabel('Important Var')
plt.title('Nice Graph')
plt.legend(loc='best')
plt.show()
```

```python
help(plt.hist)
```

#### Scatter Plots

```python
l = [(i ** 2) for i in bins]
j = [(i // 2) for i in l]

plt.scatter(bins, l, color='c', label='First Plot', s=100)
plt.scatter(bins, j, color='m', label='Second Plot', s=50, marker='x')
plt.xlabel('Plot Label')
plt.ylabel('Important Var')
plt.title('Nice Graph')
plt.legend(loc='best')
plt.show()
```

#### Stack Plots

```python
days = [i for i in range(1, 6)]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days, sleeping, eating, working, playing, colors=['c', 'm', 'r', 'y'], labels=['Sleep', 'Eat', 'Work', 'Play'])
plt.xlabel('Plot Label')
plt.ylabel('Important Var')
plt.title('Nice Graph')
plt.legend(loc='best')
plt.show()
```
