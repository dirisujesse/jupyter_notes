## Text Segmentation with Regular Expressions

<hr>
** *Segmenting the text into it's constituent words with regular expressions* **
* For this text single white spaces were the primary word boundary identifiers so we make use of that in our algorithm 

```python
# the regualar expression \s reffers to white space
words = re.split(r'\s', text) #this says split items in text by whitespace
#words = [word.replace(r'\W', '') for word in words]

[print("({})".format(word), end="|") for word in words] #this prints the whole list
print('\n')

print("{:^100}\n".format('Uneme Words'))
print("{:<20} {:>100}\n".format('S/N', 'WORD'))
#we would iterate through words which is a list object to print the items in it
for index,word in enumerate(words):
    print("{:<20} {:>100}".format(index, word))
```

<hr>
** *Segmenting the text into it's constituent sentences with regular expressions* **
* For this text fullstops were enough to signify sentence boundaries so we make use of that in our algorithm 

```python
# the regualar expression \, reffers to a fullstop
sents = re.split(r'\.', text)
for sent in sents:
    print(sent,'\n')
```

<hr><hr>
## Assigning Parts of Speech tags to the segmented words of the text
* Based on our understanding of the morpheme structure of the word classes of the language we generate regex patterns
* We feed this into an nltk method which assigns tag on the basis of our patterns

```python
#our patterns
#in regular expression | means or \b means word boundary ^ means start $ means end ? means zero or one * means 0 or mor + means one or more
patterns = [
(r'\b[AEIOUỌẸ].*[a-z]+\b', 'NNP'), #words starting with capital vowels followed by other characters are mostly proper nouns
(r'\bni\b', 'WDT'), #wh determiner (that,which) 
(r'\bma\b', 'NEG'), #negator    
(r'[\(\)\'"\.,\*\+\<=\>\?:;]', 'SYM'), #Symbols 
(r'\b(I|i).*min\b' , 'VBG' ), # words containing the circumfix I/i...min are gerunds
(r'\b[aeiouọẹ].*[a-z]+\b', 'NN'),   #words starting with vowels are mostly common nouns  
(r'^([^vzkbdfhjlnprsty]?[AEIOUaeiou]|(m)?wan|mẹ|mi|wa|e(le)?(le)?|ọ(le)?(le)?)$', 'PRP'), # a list of personalpronouns we specified the kinds of consonants pronouns cant begin with
(r'\b(ọni|eni|ọkpa)\b', 'DT'), #a list of posible determiners
(r'\b(m)?[aeiou]+la\b', 'WP'), #Wh pronouns mostly end with the -la suffix
(r'\b[bcdfghjklmnpqrstvwyz].*ni\b' , 'VBN' ), # verbs ending with -ni suffix are perfective
(r'\b(i)?k([aeiou]{1}([bcdfghjklmnpqrstvwyz]{0,2}([aeiou]{0,1}([n]{0,1})?)?)?)\b' , 'PRP$' ), # possesive pronouns
(r'\b[bcdfghjklmnprstvwyz].*\b', 'VB' ), # items begining with consonants are more likely to be verbs
(r'^-?[0-9]+(.[0-9]+)?$' , 'CD' ), # cardinal numbers
(r'.*' , 'RP' ), # particles (default)
]

regtag = nltk.RegexpTagger(patterns)
```

```python
tagged_words = regtag.tag(words)

for i in tagged_words:
    print(i, end="|")
```

## Identifying the most likely word endings, beginings and CV, VC combinations in our text

<hr>
** *Identifying the most frequent word endings in the text* **

```python
word_endings = nltk.FreqDist([word[-1] for word in text.lower().split(" ") if word.isalpha()])
#this says return the last character of a word in text split by space if word is an alphabetic letter
word_endings.plot()
#plot the frequency of the word endings
#we notice that a is the most common word ending while n is the least common, this hints at the rarity of nasal vowels in uneme
```

<hr>
** *Identifying the most frequent word endings in the text* **

```python
word_beginings = nltk.FreqDist([word[0] for word in text.lower().split(" ") if word.isalpha()])
#this says return the first character of a word in text split by space if word is an alphabetic letter
word_beginings.plot()
#plot the frequency of the word beginings
#we notice that a is the most common word beginings in our uneme text are nasals while h is the least common
```

<hr>
** *Identifying VC patterns in the text* **

```python
vc_patterns = nltk.ConditionalFreqDist([cv for word in text.lower().split(" ") 
                                        for cv in re.findall(r'[aeiou][bdfghjklmnprstvwyz]', word)])
#this says vc patterns is the list of all vowels and consonants sequences in words for words in the text split by space 
#this kind of construction is reffered to as list comprehension in python
vc_patterns.plot()
#the most common vc pattern in between a and m
vc_patterns.tabulate()
#we tabulate the patterns with vowels on the y axis and the consonants on the x axis
#we can observe that only seems to enjoy co-occuring with u
```

<hr>
** *Identifying the CV patterns in the text* **

```python
cv_patterns = nltk.ConditionalFreqDist([cv for word in text.lower().split(" ") 
                                        for cv in re.findall(r'[bdfghjklmnprstvwyz][aeiou]', word)])
#this says cv patterns is the list of all consonants and vowels sequences in words for words in the text split by space 
#this kind of construction is reffered to as list comprehension in python
cv_patterns.plot(label=False)
cv_patterns.tabulate()
#we tabulate the patterns with consonants on the y axis and the vowels on the x axis
#we can observe that only seems to enjoy co-occuring with u
#the most common pattern is beteen n and i
```
