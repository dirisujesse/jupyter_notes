```python
#required packages
import numpy as np
import pylab as pl
```

```python
#setting up an array of values representing data to be manipulates
x = [1,2,3,4,5]
y = [1,4,9,16,25]

pl.plot(x,y)

pl.show()
```

```python
#a scatter plot
pl.plot(x,y,'ro')
pl.show()
```

```python
xi = [1,2,3,4,5]
yi = [19,20,31,13,17]
pl.plot(xi,yi)
pl.show()
```

```python
xi = [1.5,2.3,3.4,4.6,5.4]
yi = [19,20,31,13,17]
pl.plot(xi,yi,'ro')
pl.show()
```

```python
#the third parameter which is a string generally serves to format theplot
#often the color of the points
#the following letters represent some of the colors representable 
#{w: white, r: red, b: blue, g: green, y: yellow, m: magenta, c: cyan, k: black}
pl.plot(xi,yi,'mo')
pl.show()
```

```python
#the line style as well can be modified
#red dashed lines
pl.plot(x,y,'r--')
pl.show()
```

```python
#green dotted lines
pl.plot(xi,yi,'g.') #using the * char would generate star lines
pl.show()
```

```python
#a list of markers
"""
’s’ square marker
’p’ pentagon marker
’*’ star marker
’h’ hexagon1 marker
’H’ hexagon2 marker
’+’ plus marker
’x’ x marker
’D’ diamond marker
’d’ thin diamond marker
"""
```

```python
#setting legends and limits
pl.title("Python Graph")
pl.xlabel("Floats")
pl.ylabel("Integers")
pl.xlim(0.0, 6.0)
pl.ylim(0, 30)
pl.plot(x,y,'+')
pl.show()
```

```python
#plotting multiple plots on the same axis
pl.plot(x,y,'r+')
pl.plot(xi,yi,'co')
pl.title("Python Graph")
pl.xlabel("Floats")
pl.ylabel("Integers")
pl.xlim(0.7, 6)
pl.show()
```

```python
#adding legends
#legends can be easily added to graphs with the pl legend method which could take up to 4 arguments
#pl.legend((a list of plots),(a list of labels),'the location of the legend in the graph', numpoints=1)
#the location of the legend are specifiable with the following keywords
#‘upper right’, ‘upper left’, ‘center’, ‘lower left’, ‘lower right’

up = [1,2,3]
down = [4,5,6]

next_ = [2,4,6]
near = [8,10,12]
pl.xlabel("Curved")
pl.ylabel("Straight")
pl.xlim(0.0, 6.4)
pl.ylim(0, 13)

pl.plot(up, down, 'r+', label="red")
pl.plot(next_, near, 'g^', label='green')

pl.legend(loc="best")
pl.show()
```

```python
#working with histograms
#the hist function is used for this purspose
data = np.random.normal(5.0,3.0,1000)
pl.hist(data)
pl.show()
```

```python
#to eliminate the border the following keyword argument may be specified
pl.hist(data, histtype='stepfilled')
pl.xlabel('data')
pl.show()
```

```python
bins = np.arange(5., 16., 1.) #somewhat equivalent to range(5,16,1)
pl.hist(bins)
pl.show()
```
